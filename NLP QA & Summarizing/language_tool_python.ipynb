```python
import language_tool_python
tool = language_tool_python.LanguageTool(language='en-US')  # 'en-US' represents American English
```

```python
# read Data
with open ("DS_MarianaTrench.txt", "r") as f:
    text = f.read()
```

```python
#text='Hello World!'
matches = tool.check(text)
```

```python
if len(matches) == 0:
    print("Paragraph is correct")
else:
    for match in matches:
        print("Error:", match.ruleId)
        print("Message:", match.message)
        print("Replacements:", match.replacements)
        print("-----------")
```
